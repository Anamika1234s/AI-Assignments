```python
# Convert date of joining into datetime

df['Date of Joining']=pd.to_datetime(df['Date of Joining'],dayfirst=True)
```

```python
#1. Greet employee with their first name

def Greetings(EID):
    emp= df[df['EID'] == EID]
    if not emp.empty:
        name=emp.iloc[0]["Employee Name"]
        first_name = name.split()[0]
        gender = emp.iloc[0]["Gender"]
        title = "Mr" if gender.lower().startswith("m") else "Ms"
        return f"Hello, {title} {first_name}"
    return "Employee Not Found"
            
while True:
    emp_id = input("Enter Employee ID (or 'exit' to quit): ")
    if emp_id.lower() == "exit":
        print("Exiting...")
        break
    print(Greetings(emp_id))
    

```

```python
#2.Function for calculating the average salary of each department
def average_salary_by_department(DataFrame):
    avgsalary=DataFrame.groupby("Department")["Salary"].mean()
    return avgsalary

print(average_salary_by_department(df))
```

```python
#3. Function to display employee details by ID
def get_employee_details(emp_id, dataframe):
    emp = dataframe[dataframe["EID"] == emp_id]
    if not emp.empty:
        name = emp.iloc[0]["Employee Name"]
        doj = emp.iloc[0]["Date of Joining"]
        dept = emp.iloc[0]["Department"]
        return f"{name} | {doj} | {dept}"
    else:
        return "Employee ID not found."


while True:
    emp_id = input("Enter Employee ID (or 'exit' to quit): ")
    if emp_id.lower() == "exit":
        print("Exiting...")
        break
    print(get_employee_details(emp_id, df))
```

```python
#4.Function for sorting employee from oldest to newest.
def sort_employee_by_joining():
    sorted_df=df.sort_values(by="Date of Joining", ascending=True)
    return sorted_df[["EID", "Employee Name", "Date of Joining", "Department"]]

print(sort_employee_by_joining())
```
